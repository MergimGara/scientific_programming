```python
import numpy as np
import matplotlib.pyplot as plt
import os
import platform
import socket
from platform import python_version
from datetime import datetime

# Create Numpy Array with Random Numbers
# Erstellen Sie ein NumPy-Array mit zufälligen Zahlen.

# Create a numpy array with random numbers
random_numbers = np.random.rand(100)  # Array with 100 random numbers between 0 and 1
# Calculate Quartiles of the Random Numbers
# Berechnen Sie die Quartile der zufälligen Zahlen mit der Methode percentile() aus der NumPy-Bibliothek.

# Calculate the quartiles
quartiles = np.percentile(random_numbers, [25, 50, 75])

# Display the quartiles
quartiles

plt.hist(random_numbers, bins=10, edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.boxplot(random_numbers)
plt.title('Boxplot of Random Numbers')
plt.ylabel('Value')
plt.show()

print('-----------------------------------')
print(os.name.upper())
print(platform.system(), '|', platform.release())
print('Datetime:', datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
print('Python Version:', python_version())
print('-----------------------------------')
```
