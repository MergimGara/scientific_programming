```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Fiktive Daten erstellen
data = {
    'Therapie': ['Traditionelle Physiotherapie', 'Traditionelle Physiotherapie', 'Traditionelle Physiotherapie', 'KI-gestützte Physiotherapie', 'KI-gestützte Physiotherapie', 'KI-gestützte Physiotherapie'],
    'Parameter': ['Rückfallquote', 'Behandlungsdauer', 'Therapieabbrüche', 'Rückfallquote', 'Behandlungsdauer', 'Therapieabbrüche'],
    'Wert': [30, 60, 20, 10, 30, 5]  # Fiktive Werte
}

df = pd.DataFrame(data)

# Graphen erstellen
plt.figure(figsize=(12, 6))
sns.barplot(x='Parameter', y='Wert', hue='Therapie', data=df)
plt.title('Vergleich der traditionellen vs. KI-gestützten Physiotherapie')
plt.ylabel('Prozent / Tage')
plt.xlabel('Parameter')
plt.legend(title='Therapieform')
plt.show()
```
